```python
import os
import random
import shutil

# Set paths
base_path = '/content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection'
images_path = os.path.join(base_path, 'images')
labels_path = os.path.join(base_path, 'labels')

# Create new folders
for split in ['train', 'val', 'test']:
    os.makedirs(f"{base_path}/{split}/images", exist_ok=True)
    os.makedirs(f"{base_path}/{split}/labels", exist_ok=True)

# Get all image files
all_images = [f for f in os.listdir(images_path) if f.endswith('.jpg')]
random.shuffle(all_images)

# 70% train, 15% val, 15% test
train_count = int(len(all_images) * 0.8)
train_images = all_images[:train_count]
val_images = all_images[train_count:]

# Function to copy images and labels
def copy_files(image_list, split):
    for image in image_list:
        shutil.copy(os.path.join(images_path, image), os.path.join(base_path, split, 'images', image))
        label_file = image.replace('.jpg', '.txt')
        shutil.copy(os.path.join(labels_path, label_file), os.path.join(base_path, split, 'labels', label_file))

copy_files(train_images, 'train')
copy_files(val_images, 'val')

# Calculate splits
train_count = int(len(all_images) * 0.7)
val_count = int(len(all_images) * 0.15)
test_count = len(all_images) - train_count - val_count

train_images = all_images[:train_count]
val_images = all_images[train_count:train_count + val_count]
test_images = all_images[train_count + val_count:]

# Copy all sets
copy_files(train_images, 'train')
copy_files(val_images, 'val')
copy_files(test_images, 'test')
```

```python
data_yaml = """
train: /content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection/train/images
val: /content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection/val/images
test: /content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection/test/images

nc: 1
names:
  - arecanut
"""

with open('/content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection/data.yaml', 'w') as file:
    file.write(data_yaml)
```

```python
# Load model (use yolov8n.pt for a light version)
model = YOLO('yolov8n.pt')

# Train
model.train(data=f'{base_path}/data.yaml', epochs=50, imgsz=640)
```

```python
results = model.train(
    data='/content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection/data.yaml',  # Path to data.yaml
    epochs=100,  # Number of epochs for training
    imgsz=640,   # Image size
    project='/content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection',  # Parent folder for results
    name='YOLO_training_results',  # Subfolder name
    exist_ok=True  # Overwrite if folder already exists
)
```

```python
results = model.val(
    data='/content/drive/MyDrive/Arecanut_Detection_with_ML/Arecanut_Detection/YOLO_Detection/data.yaml',
    split='test'  # This tells YOLO to use the test set
)
```
