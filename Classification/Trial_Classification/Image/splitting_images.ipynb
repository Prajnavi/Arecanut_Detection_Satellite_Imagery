```python
import os
import cv2
from google.colab import drive
```

```python
# Load the saved Google Earth image
INPUT_FOLDER = "/content/gdrive/MyDrive/Arecanut_Detection_with_ML/Classification/Trial_Classification/Image/Images"
```

```python
# Define output
OUTPUT_FOLDER = "/content/gdrive/MyDrive/Arecanut_Detection_with_ML/Classification/Trial_Classification/splitted_images"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
# Estimated meters per pixel
METERS_PER_PIXEL = 1  # Example: If 200m = 400 pixels, then MPP = 0.5
TILE_SIZE = int(200 / METERS_PER_PIXEL)  # Convert 200m to pixels
#TILE_SIZE =200
```

```python
# Loop through all image files in the folder
for filename in os.listdir(INPUT_FOLDER):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif')):
        IMAGE_PATH = os.path.join(INPUT_FOLDER, filename)
        image = cv2.imread(IMAGE_PATH)

        if image is None:
            print(f"Error: Could not load image from {IMAGE_PATH}. Please check the file path and permissions.")
            continue
        else:
            h, w, _ = image.shape

            # Split image into tiles
            tile_count = 0
            base_name = os.path.splitext(filename)[0]

            for y in range(0, h, TILE_SIZE):
                for x in range(0, w, TILE_SIZE):
                    tile = image[y:y+TILE_SIZE, x:x+TILE_SIZE]

                    # Ensure the tile is within image bounds
                    if tile.shape[0] == TILE_SIZE and tile.shape[1] == TILE_SIZE:
                        tile_filename = f"{base_name}_tile_{tile_count}.png"
                        tile_path = os.path.join(OUTPUT_FOLDER, tile_filename)
                        cv2.imwrite(tile_path, tile)
                        tile_count += 1

            print(f"Saved {tile_count} sub-images from '{filename}' in '{OUTPUT_FOLDER}'")
```
