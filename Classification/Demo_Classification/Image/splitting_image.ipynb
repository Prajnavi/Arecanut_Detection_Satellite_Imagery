```python
# Load the saved Google Earth image
IMAGE_PATH = "/content/gdrive/MyDrive/Arecanut_Detection_with_ML/Classification/Demo_Classification/Image/kansuru_image3.jpg"
```

```python
# Define output foldermageprojects
OUTPUT_FOLDER = "/content/gdrive/MyDrive/Arecanut_Detection_with_ML/Classification/Demo_Classification/splitted_images"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
# Estimated meters per pixel (UPDATE based on your measurement)
METERS_PER_PIXEL = 1  # Example: If 200m = 400 pixels, then MPP = 0.5
TILE_SIZE = int(200 / METERS_PER_PIXEL)  # Convert 200m to pixels
```

```python
# Load the image
image = cv2.imread(IMAGE_PATH)
if image is None:
    print(f"Error: Could not load image from {IMAGE_PATH}. Please check the file path and permissions.")
else:
    h, w, _ = image.shape
```

```python
h, w, _ = image.shape

# Split image into tiles
tile_count = 0
for y in range(0, h, TILE_SIZE):
    for x in range(0, w, TILE_SIZE):
        tile = image[y:y+TILE_SIZE, x:x+TILE_SIZE]

        # Ensure the tile is within image bounds
        if tile.shape[0] == TILE_SIZE and tile.shape[1] == TILE_SIZE:
            tile_path = os.path.join(OUTPUT_FOLDER, f"tile_{tile_count}.png")
            cv2.imwrite(tile_path, tile)
            tile_count += 1

print(f"Saved {tile_count} sub-images in '{OUTPUT_FOLDER}'")
```
